```python
class Node:
    def __init__(self, data):
        self.data = data
        self.leftNode = None
        self.rightNode = None

class BinaryTree:
    def __init__(self, preOrder, inOrder):
        self.root = self.Tree(preOrder, inOrder)

    def Tree(self, preOrder, inOrder):
        if len(preOrder)==0: # 確認前序是否為空
            return None
        root=Node(preOrder[0]) # 找根節點
        TinIndex=inOrder.index(preOrder[0]) # 找根節點在中序的位子
        root.leftNode=self.Tree(preOrder[1:TinIndex+1], inOrder[0:TinIndex]) # 判斷根節點左邊
        root.rightNode=self.Tree(preOrder[TinIndex+1:], inOrder[TinIndex+1:]) # 判斷根節點右邊
        return root
    def postOrder(self, curNode):  
        data = ''

        if None != curNode:
            data += self.postOrder(curNode.leftNode)
            data += self.postOrder(curNode.rightNode)
            data += curNode.data

        return data
```

```python
order = {
    'PreOrder': ['ABDEFG', 'ABHJCDFGE', 'ACGDHBEIF', 'A', 'AB', 'AB', 'ABC', 'ABC', 'ABC', 'ABCD'],
    'InOrder': ['BAEDGF', 'HBJAFDGCE', 'GCHDAIEFB', 'A', 'BA', 'AB', 'BAC', 'CBA', 'ABC', 'CBDA'],
    'PostOrder': ['BEGFDA', 'HJBFGDECA', 'GHDCIFEBA', 'A', 'BA', 'BA', 'BCA', 'CBA', 'CBA', 'CDBA']
}

for index in range(len(order['PreOrder'])):
    binaryTree = BinaryTree(list(order['PreOrder'][index]), list(order["InOrder"][index]))
    data = binaryTree.postOrder(binaryTree.root)

    print('-' * 30)
    print('Input')
    print(f'PreOrder: {order["PreOrder"][index]}')
    print(f'InOrder: {order["InOrder"][index]}\n')
    print('Output')
    print(f'PostOrder: {data}')
    print(f'Checkpoint: {data == order["PostOrder"][index]}')

    del binaryTree
```
