```python
#double linked list
class Node:
    def __init__(self, name=None, amount=None, date=None):
        self.name = name
        self.amount = amount
        self.date = date
        self.preNode = None
        self.nextNode = None

class DoublyLinkedList:
    def __init__(self):
        self.headNode = None
        self.lastNode = None

    def append(self, name, amount, date):
        newNode = Node(name, amount, date)

        if None == self.headNode:
            self.headNode = newNode
        else:
            newNode.preNode = self.lastNode
            self.lastNode.nextNode = newNode

        self.lastNode = newNode

    def insert(self, name, inserName, insertamount, date):
        if self.headNode == None:
            self.headNode = Node(inserName,insertamount,date)
            self.lastNode = self.headNode
        else:
            currentNode = Node()
            currentNode = self.headNode
            tempNode = Node(inserName,insertamount,date)
            if self.lastNode.name == name:
               self.append(inserName, insertamount, date)
            else:
                while currentNode.nextNode != None:
                    if currentNode.name == name:
                        tempNode.nextNode = currentNode.nextNode
                        tempNode.nextNode.preNode = tempNode
                        tempNode.preNode = currentNode
                        currentNode.nextNode = tempNode
                    currentNode = currentNode.nextNode


    def update(self,name, amount, date):
        if self.headNode == None:
            print("NO DATA")
        else:
            currentNode = Node()
            currentNode = self.headNode
            counter = 0
            while currentNode != None:
                if currentNode.name == name:
                    currentNode.date = date
                    currentNode.amount = amount
                    counter += 1
                currentNode = currentNode.nextNode
            if counter == 0:
                print("Not Found")

    def delete(self, name):
        if self.headNode == None:
            print("NO DATA")
        currentNode = Node()
        currentNode = self.headNode
        pre = Node()
        pre = self.headNode
        while currentNode != None :
            pre=currentNode
            currentNode=currentNode.nextNode
            if pre.name==name:
                if currentNode == None:
                    pre.preNode.nextNode = pre.nextNode
                    self.lastNode = self.lastNode.preNode
                else:
                    if pre.preNode == None:
                        pre.nextNode.preNode = pre.preNode
                        self.headNode = self.headNode.nextNode
                    else:
                        pre.nextNode.preNode = pre.preNode
                        pre.preNode.nextNode = pre.nextNode
                        pre = None


    def showForward(self):
        currentNode = self.headNode

        print('\n' + '-' * 100)
        print('Forward')

        while None != currentNode:
            print(f'name: {currentNode.name}\namount: {currentNode.amount}\ndate: {currentNode.date}\n')
            currentNode = currentNode.nextNode

        print('-' * 100)

    def showBackward(self):
        currentNode = self.lastNode

        print('-' * 100)
        print('Backward')

        while None != currentNode:
            print(f'name: {currentNode.name}\namount: {currentNode.amount}\ndate: {currentNode.date}\n')
            currentNode = currentNode.preNode

        print('-' * 100 + '\n')

    def showData(self):
        self.showForward()
        self.showBackward()
```

```python

from datetime import datetime

doublyLinkedList = DoublyLinkedList()

def getDate(year, month, day, hours, minute):
    return datetime(year, month, day, hours, minute).strftime('%Y-%m-%d(%a) %H:%M')

data = {
    'Wayne': {'amount': 3, 'date': getDate(2022, 10, 13, 20, 00)},
    'Kim': {'amount': 4, 'date': getDate(2022, 10, 15, 20, 00)},
    'Nina': {'amount': 1, 'date': getDate(2022, 11, 1, 20, 00)},
    'John': {'amount': 2, 'date': getDate(2022, 10, 13, 19, 30)},
    'Lucia': {'amount': 5, 'date': getDate(2022, 12, 31, 21, 30)},
}

for name in data:
    doublyLinkedList.append(name, data[name]['amount'], data[name]['date'])
doublyLinkedList.showData()

print('Insert Tom')
doublyLinkedList.insert('Nina', 'Tom', 8, getDate(2022, 10, 13, 19, 30))
doublyLinkedList.showData()

print('Insert Antony')
doublyLinkedList.insert('Lucia', 'Antony', 7, getDate(2022, 10, 12, 12, 30))
doublyLinkedList.showData()

print('Delete Wayne')
doublyLinkedList.delete('Wayne')
doublyLinkedList.showData()

print('Delete Nina')
doublyLinkedList.delete('Nina')
doublyLinkedList.showData()

print('Delete Lucia')
doublyLinkedList.delete('Lucia')
doublyLinkedList.showData()

print('Update Lucia')
doublyLinkedList.update('Lucia', 2, getDate(2022, 12, 31, 20, 00))
doublyLinkedList.showData()

print('Update Tom')
doublyLinkedList.update('Tom', 6, getDate(2022, 10, 12, 20, 00))
doublyLinkedList.showData()
```
